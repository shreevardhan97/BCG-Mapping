```python
#import school from SANGRUR folder
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read all csv from SANGRUR folder and merge them
import os
path = './SANGRUR'
files = os.listdir(path)
files_csv = [f for f in files if f[-3:] == 'csv']
df = pd.DataFrame()
for f in files_csv:
    data = pd.read_csv(os.path.join(path, f))
    df = df.append(data)
    #remove duplicate
df = df.drop_duplicates()
df.head()
# show all columns
pd.set_option('display.max_columns', None)
df.head()
#merge the blocks and df
#read excel
df_block = pd.read_excel('./blocks/LUDHIANA.xlsx')
df_block.head()
#make first row as header
new_header = df_block.iloc[0] #grab the first row for the header
df_block = df_block[1:] #take the data less the header row
df_block.columns = new_header #set the header row as the df header
df_block.head()
```

```python
df.head()
```

```python
# merge df and df_block and remove using UDISE and SCHCD
df_master = pd.merge(df, df_block, left_on='features__attributes__schcd', right_on='UDISE', how='inner')
#check number of misses for ludhiana
```

```python
df_master.head()
#total number of rows
df_master.shape
df_master.head()
# df master should only have UDISE lat long and school name
df_master = df_master[['UDISE', 'features__attributes__latitude', 'features__attributes__longitude', 'School Name']]
df_master.drop_duplicates(inplace=True)
```
